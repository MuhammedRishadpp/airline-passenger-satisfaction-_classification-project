```python
# input data
x=df.iloc[:,:-1].values
x
```

```python
# output data
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python

from sklearn.neighbors import KNeighborsClassifier
neighbours_model=KNeighborsClassifier(n_neighbors=5)
```

```python
from sklearn.svm import SVC
svm_model=SVC()
```

```python

from sklearn.tree import DecisionTreeClassifier
tree_model=DecisionTreeClassifier()
```

```python

from sklearn.naive_bayes import GaussianNB
naive_bayes_model=GaussianNB()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay,classification_report
model=[neighbours_model,svm_model,tree_model,naive_bayes_model]
for i in model:
  i.fit(x_train,y_train)
  y_prede=i.predict(x_test)
  result=confusion_matrix(y_test,y_prede)
  score=accuracy_score(y_test,y_prede)
  report=classification_report(y_test,y_prede)
  cm=['neutral or dissatisfied', 'satisfied']
  cmd=ConfusionMatrixDisplay(result,display_labels=cm)
  print("model name :- ",i,"\n",y_prede)
  print(result)
  print(score)
  print(report)
  print(cmd.plot())
  print("*"*100)
```
